```python
import pandas as pd
```

```python
deathCause = pd.read_excel('ori_data/사망원인_104항목__성_연령_5세_별_사망자수__사망률_20230721170151.xlsx')
```

```python
deathCause.info()

# 총 1141개 인덱스 중 
# 사망원인, 성별 컬럼에 결측치 존재
# 결측치를 drop() 하면 해당 행이 지워지기 때문에 결측치를 치환하는 방향으로 결측치 처리
# 심장질환 데이터의 연령 컬럼이 10세 단위로 구성됨. 떄문에, 사망원인 데이터의 연령 컬럼은 10세 단위로 변경이 필요함
```

```python
# 인덱스 0 삭제
deathCause.drop(deathCause.index[0], inplace=True)

# 인덱스 번호 리셋
deathCause.reset_index(drop=True, inplace=True)
```

```python
# 컬럼 설명하는 행 삭제 후 데이터 복사
deathCause_copy = deathCause.copy()
```

```python
deathCause_copy
```

```python
def fill_missing_cause_of_death(df):
    cause_of_death = None  # 초기 사망원인 초기화
    
    for idx, row in df.iterrows():  # 데이터프레임의 각 행에 대한 반복
        if row['사망원인별(104항목)'] == '계':
            cause_of_death = '계'  # '계'인 경우 cause_of_death 변수에 '계' 할당
        elif pd.notnull(row['사망원인별(104항목)']):
            cause_of_death = row['사망원인별(104항목)']  # 결측치가 아니면 현재 값 할당
        
        df.loc[idx, '사망원인별(104항목)'] = cause_of_death  # 원본 데이터프레임에 값을 할당

    return df  # 수정된 데이터프레임 반환

# deathCause_copy 데이터프레임을 함수에 전달하여 결측치를 채움
fill_missing_cause_of_death(deathCause_copy)
```

```python
def fill_missing_sex(df):
    gender = None  # 초기 성별 초기화
    
    for idx, row in df.iterrows():  # 데이터프레임의 각 행에 대한 반복
        if row['성별'] == '계':
            gender = '계'  # '계'인 경우 gender 변수에 '계' 할당
        elif row['성별'] == '남자':
            gender = '남자'  # '남자'인 경우 gender 변수에 '남자' 할당
        elif row['성별'] == '여자':
            gender = '여자'  # '여자'인 경우 gender 변수에 '여자' 할당
        
        # gender 변수에 따라 성별 값을 할당
        df.loc[idx, '성별'] = gender

    return df  # 수정된 데이터프레임 반환

# deathCause_copy 데이터프레임을 함수에 전달하여 결측치를 채움
fill_missing_sex(deathCause_copy)
```

```python
def modify_age_ranges(df):
    for idx, row in df.iterrows():    # 데이터프레임의 각 행에 대한 반복
        age_range = row['연령(5세)별']
        if age_range in ['0세', '1 - 4세', '5 - 9세']:
            df.loc[idx, '연령(5세)별'] = '0 - 9세'
        elif age_range in ['10 - 14세', '15 - 19세']:
            df.loc[idx, '연령(5세)별'] = '10 - 19세'
        elif age_range in ['20 - 24세', '25 - 29세']:
            df.loc[idx, '연령(5세)별'] = '20 - 29세'
        elif age_range in ['30 - 34세', '35 - 39세']:
            df.loc[idx, '연령(5세)별'] = '30 - 39세'
        elif age_range in ['40 - 44세', '45 - 49세']:
            df.loc[idx, '연령(5세)별'] = '40 - 49세'
        elif age_range in ['50 - 54세', '55 - 59세']:
            df.loc[idx, '연령(5세)별'] = '50 - 59세'
        elif age_range in ['60 - 64세', '65 - 69세']:
            df.loc[idx, '연령(5세)별'] = '60 - 69세'
        elif age_range in ['70 - 74세', '75 - 79세']:
            df.loc[idx, '연령(5세)별'] = '70 - 79세'
        elif age_range == '80세이상':
            df.loc[idx, '연령(5세)별'] = '80세 이상'
    return df
    

modify_age_ranges(deathCause_copy) 
```

```python
df = deathCause_copy.groupby(['사망원인별(104항목)', '성별', '연령(5세)별'], sort=False).sum().reset_index()
```
