## 데이터 구성
- HeartDisease: 심장질환 여부 (No: 심장질환 없음, Yes: 심장질환 있음)
- BMI: 체질량 지수
- Smoking: 흡연 여부 (Yes: 흡연, No: 비흡연)
- AlcoholDrinking: 음주 여부 (No: 미음주, Yes: 음주)
- Stroke: 뇌졸중 발병 여부 (Yes: 발병, No: 미발병)
- PhysicalHealth: 신체 건강 상태 (범주형)
- MentalHealth: 정신 건강 상태 (범주형)
- DiffWalking: 걷기 어려움 정도 (No: 없음, Yes: 있음)
- Sex: 성별 (Male : 남성, Female: 여성)
- AgeCategory: 나이 범주
- Race: 인종 ('White', 'Black', 'Asian', 'American Indian/Alaskan Native',
       'Other', 'Hispanic')
- Diabetic: 당뇨병 여부 (Yes: 있음, No: 없음, No, borderline diabetes: 당뇨병 전단계, 'Yes: 임신 중 당뇨병 여부)
- PhysicalActivity: 신체 활동 (Yes: 신체 활동 유, No: 신체 활동 무)
- GenHealth: 일반적인 건강 상태 (Poor, Fair, Very good, Excellent)
- SleepTime: 수면 시간 (1~24시간 이상)
- Asthma: 천식 발병 여부 (Yes: 발병, No: 미발병)
- KidneyDisease: 신장 질환 여부 (Yes: 있음, No: 없음)
- SkinCancer: 피부암 발병 여부 (Yes:발병, No: 미발병)

```python
# 모든 범주형 열을 가져옵니다.
columns = df.select_dtypes(include=['object']).columns.tolist()
columns = columns[1:]  # 첫 번째 열은 'HeartDisease' 이므로 제외합니다.

# subplot의 개수를 계산합니다.
num_plots = len(columns)

# 행과 열의 개수를 계산합니다. (예: 3개의 범주형 열 -> 2x2 그리드)
cols = 2  # 열의 개수 (조정 가능)
rows = (num_plots + cols - 1) // cols

# 그리드를 생성하고 그림의 세로 크기를 늘립니다.
fig, axes = plt.subplots(rows, cols, figsize=(12, 4*rows))

# 범주형 열을 순회하면서 subplot에 카운트 플롯을 그립니다.
for i, column in enumerate(columns):
    if column == 'HeartDisease':  # 'HeartDisease'와 'Race' 열은 제외
        continue
    row = i // cols  # 행 인덱스 계산
    col = i % cols   # 열 인덱스 계산
    ax = axes[row, col]  # 현재 subplot을 선택
    
    if column == 'GenHealth' or column == 'AgeCategory':
        if column == 'GenHealth':
            custom_order = ['Poor', 'Fair', 'Good', 'Very good', 'Excellent']
        elif column == 'AgeCategory':
            custom_order = sorted(df['AgeCategory'].unique())
        sns.countplot(data=df, x=column, hue='HeartDisease', ax=ax, order=custom_order)
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45) # 겹침으로 인한 각도 변경
    else:
        sns.countplot(data=df, x=column, hue='HeartDisease', ax=ax)
    
    ax.set_title(f'{column} vs. HeartDisease')  # subplot 제목 설정

# 나머지 subplot을 숨깁니다.
for i in range(num_plots, rows * cols):
    fig.delaxes(axes.flatten()[i])

plt.tight_layout()  # subplot 간의 간격을 조절하여 레이아웃을 보정
plt.show()
```

```python
def circle_chart(feature):
    labels = ['White','Hispanic','Black','Other','Asian','American Indian/Alaskan Native']
    race = df[feature].value_counts()
    plt.figure(figsize=(5,5))
    plt.pie(race,labels=labels,autopct='%2.1f%%',shadow=True)
    circle = plt.Circle(xy  = (0,0),radius = 0.65,facecolor='white')
    plt.gca().add_artist(circle)
    plt.legend(title='Race',loc='upper right',framealpha=0.5,fontsize=8,bbox_to_anchor=(2,1))
    plt.show()
```

## 설명
- 주어진 데이터의 특성을 0과 1 사이의 값으로 스케일링하는 데 사용된다.
- 이러한 스케일링을 "최소-최대 스케일링" 또는 "정규화(Normalization)"라고도 한다.

- MinMaxScaler를 사용하면 주로 다음과 같은 상황에서 도움이 된다.

다양한 단위나 범위를 가진 특성이 있다. 데이터의 특성을 동일한 범위로 맞춰주어 각 특성을 공평하게 고려한다.

일부 모델 알고리즘(예: 경사 하강법 기반 알고리즘)에서 스케일링이 중요하다. 특성 스케일링은 모델의 수렴 속도와 성능에 영향을 줄 수 있다.

이상치가 있는 데이터에서도 최소-최대 스케일링은 이상치에 민감하지 않다. 이 방법은 데이터의 최솟값과 최댓값을 고정시켜 이상치의 영향을 최소화한다.

MinMaxScaler는 주로 선형 회귀, 로지스틱 회귀, 신경망과 같은 알고리즘에서 사용된다. 이 모델들은 입력 데이터의 스케일에 민감하므로 스케일링이 필수이다.

주의할 점은 학습 데이터와 동일한 스케일러를 사용하여 테스트 데이터를 변환해야 한다.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC

def model(X_train, Y_train, X_test, Y_test):
    model_name = ['LogisticRegression', 'KNNClassification', 'AdaBoostClassifier', 'DecisionTreeClassifier', 'XGBClassifier', 'SVM', 'RandomForestClassifier']
    accuracy = []
    train_accuracy = []

    # LogisticRegression
    log = LogisticRegression()
    log.fit(X_train, Y_train)
    accuracy.append(log.score(X_test, Y_test))
    train_accuracy.append(log.score(X_train, Y_train))

    # KNNClassification
    neigh = KNeighborsClassifier()
    neigh.fit(X_train, Y_train)
    accuracy.append(neigh.score(X_test, Y_test))
    train_accuracy.append(neigh.score(X_train, Y_train))

    # AdaBoostClassification
    ada = AdaBoostClassifier()
    ada.fit(X_train, Y_train)
    accuracy.append(ada.score(X_test, Y_test))
    train_accuracy.append(ada.score(X_train, Y_train))

    # DecisionTreeClassifier
    clf = DecisionTreeClassifier()
    clf.fit(X_train, Y_train)
    accuracy.append(clf.score(X_test, Y_test))
    train_accuracy.append(clf.score(X_train, Y_train))

    # XGBClassifier
    xgb = XGBClassifier()
    xgb.fit(X_train, Y_train)
    accuracy.append(xgb.score(X_test, Y_test))
    train_accuracy.append(xgb.score(X_train, Y_train))

    # SVM
    svm = SVC()
    svm.fit(X_train, Y_train)
    accuracy.append(svm.score(X_test, Y_test))
    train_accuracy.append(svm.score(X_train, Y_train))

    # Random Forest Classifier
    rt = RandomForestClassifier()
    rt.fit(X_train, Y_train)
    accuracy.append(rt.score(X_test, Y_test))
    train_accuracy.append(rt.score(X_train, Y_train))

    return model_name, accuracy, train_accuracy

# 함수 호출 및 결과 출력
model_name, accuracy, train_accuracy = model(X_train, Y_train, x_test, y_test)

for name, acc, train_acc in zip(model_name, accuracy, train_accuracy):
    print(f'Model: {name}')
    print(f'Test Accuracy: {acc}')
    print(f'Train Accuracy: {train_acc}')
    print()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score, recall_score

def model(X_train, Y_train, X_test, Y_test):
    model_name = ['LogisticRegression', 'KNNClassification', 'AdaBoostClassifier', 'DecisionTreeClassifier', 'XGBClassifier', 'SVM', 'RandomForestClassifier']
    accuracy = []
    train_accuracy = []
    f1_scores = []  # f1-score를 저장할 리스트 추가
    recalls = []    # recall을 저장할 리스트 추가

    # LogisticRegression
    log = LogisticRegression()
    log.fit(X_train, Y_train)
    accuracy.append(log.score(X_test, Y_test))
    train_accuracy.append(log.score(X_train, Y_train))
    y_pred = log.predict(X_test)
    f1_scores.append(f1_score(Y_test, y_pred))
    recalls.append(recall_score(Y_test, y_pred))

    # KNNClassification
    neigh = KNeighborsClassifier()
    neigh.fit(X_train, Y_train)
    accuracy.append(neigh.score(X_test, Y_test))
    train_accuracy.append(neigh.score(X_train, Y_train))
    y_pred = neigh.predict(X_test)
    f1_scores.append(f1_score(Y_test, y_pred))
    recalls.append(recall_score(Y_test, y_pred))

    # AdaBoostClassifier
    ada = AdaBoostClassifier()
    ada.fit(X_train, Y_train)
    accuracy.append(ada.score(X_test, Y_test))
    train_accuracy.append(ada.score(X_train, Y_train))
    y_pred = ada.predict(X_test)
    f1_scores.append(f1_score(Y_test, y_pred))
    recalls.append(recall_score(Y_test, y_pred))

    # DecisionTreeClassifier
    clf = DecisionTreeClassifier()
    clf.fit(X_train, Y_train)
    accuracy.append(clf.score(X_test, Y_test))
    train_accuracy.append(clf.score(X_train, Y_train))
    y_pred = clf.predict(X_test)
    f1_scores.append(f1_score(Y_test, y_pred))
    recalls.append(recall_score(Y_test, y_pred))

    # XGBClassifier
    xgb = XGBClassifier()
    xgb.fit(X_train, Y_train)
    accuracy.append(xgb.score(X_test, Y_test))
    train_accuracy.append(xgb.score(X_train, Y_train))
    y_pred = xgb.predict(X_test)
    f1_scores.append(f1_score(Y_test, y_pred))
    recalls.append(recall_score(Y_test, y_pred))

    # SVM
    svm = SVC()
    svm.fit(X_train, Y_train)
    accuracy.append(svm.score(X_test, Y_test))
    train_accuracy.append(svm.score(X_train, Y_train))
    y_pred = svm.predict(X_test)
    f1_scores.append(f1_score(Y_test, y_pred))
    recalls.append(recall_score(Y_test, y_pred))

    # RandomForestClassifier
    rt = RandomForestClassifier()
    rt.fit(X_train, Y_train)
    accuracy.append(rt.score(X_test, Y_test))
    train_accuracy.append(rt.score(X_train, Y_train))
    y_pred = rt.predict(X_test)
    f1_scores.append(f1_score(Y_test, y_pred))
    recalls.append(recall_score(Y_test, y_pred))

    return model_name, accuracy, train_accuracy, f1_scores, recalls

# 함수 호출 및 결과 출력
model_name, accuracy, train_accuracy, f1_scores, recalls = model(X_train, Y_train, x_test, y_test)

for name, acc, train_acc, f1, recall in zip(model_name, accuracy, train_accuracy, f1_scores, recalls):
    print(f'Model: {name}')
    print(f'Test Accuracy: {acc}')
    print(f'Train Accuracy: {train_acc}')
    print(f'F1 Score: {f1}')
    print(f'Recall: {recall}')
    print()
```
